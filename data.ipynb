STUDENT MARKS PREDICTION MODEL



The Linear Regression model achieved an R² score of 0.81, meaning the model explains about 81% of the variation in students' final grades (G3). Features like G1 and G2 showed strong predictive power, indicating previous academic performance is the most important factor affecting final result


The Mean Absolute Error shows that the predicted final grade differs from the actual grade by approximately 1.29 marks on average, indicating good prediction accuracy of the model.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

```

```python
df=pd.read_csv("Student_data.csv")
```

```python
features=['age','traveltime','freetime','health','studytime','activities','internet','absences','G1','G2','failures']
df[features].describe()
```

```python
le=LabelEncoder()
df['internet']=le.fit_transform(df['internet'])
```

```python
x=df[features]
y=df.G3
```

```python
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#features=['age','traveltime','freetime','health','studytime','activities','internet','absences','G1','G2','failures']

model.predict([[13,2,1,2,8,1,1,0,10,15,0]])
```

```python
pred=model.predict(x_val)
acc=r2_score(y_val,pred)
print(acc*100)
mae=mean_absolute_error(y_val,pred)
print(mae)
mse=mean_squared_error(y_val,pred)
print(mse)
```
